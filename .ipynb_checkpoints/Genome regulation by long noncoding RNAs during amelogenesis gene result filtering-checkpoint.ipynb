```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import networkx as nx
from itertools import combinations
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
#import all data
ppi_file = "data/10116.protein.actions.v11.0.txt"
ppi = pd.read_csv(ppi_file, sep='\t', header=(0))
data_files = {"AntiMS" : "data/AntiMS.tab",
              "AntiSC" : "data/AntiSC.tab",
              "LincMS" : "data/LincRNA_MS.tab",
              "LincSC" : "data/LincRNA_SC.tab"}
datas = {}
for i in data_files:
    datas[i] = pd.read_csv(data_files[i], sep='\t', header=(0))
datas['MS'] = pd.concat([datas["AntiMS"], datas["LincMS"]])
datas['SC'] = pd.concat([datas["AntiSC"], datas["LincSC"]])
STRING_mapping = pd.read_csv("data/genes_mapped", sep='\t', header=None)
STRING_mapping.columns = ['gene', 'STRING_id']
#STRING_mapping
for i in datas:
    print(i, len(datas[i]))
datas['MS'].to_csv('data/MS.csv', index = None, header=True)
datas['SC'].to_csv('data/SC.csv', index = None, header=True)
```

```python
#plot gene networ distribution
sns.distplot(ppi['score'])
```

```python
#construct network for calculation
G = nx.Graph()

all_edge = 7726910
count = 0
for idx, r in ppi.iterrows():
    if r['score'] <= 0:continue
    G.add_edge(r['item_id_a'], r['item_id_b'])
    if count % 2000000 == 0:print(count/all_edge * 100, end ="% ")
    count += 1
```

```python
#map transcript IDs into STRING network node IDs
def mapping_gene(genes):
    mapped = pd.merge(genes, STRING_mapping , left_on='#NearbyGene', right_on='gene')['STRING_id']
    print(len(mapped))
    return list(set(mapped))

print("mapping AntiSC genes")
antisc_genes = mapping_gene(datas['AntiSC'])
print("mapping SC genes")
sc_genes = mapping_gene(datas['SC'])
print("mapping AntiMS genes")
antims_genes = mapping_gene(datas['AntiMS'])
print("mapping MS genes")
ms_genes = mapping_gene(datas['MS'])
```

```python
#compute distance between each pair
all_mapped_genes = set(antisc_genes + sc_genes + antims_genes + ms_genes)
comb = combinations(all_mapped_genes, 2)
dis_m = {k:{} for k in all_mapped_genes}
for pair in comb:
    s, t = pair
    if s in G and t in G:
        try:
            dis = nx.shortest_path_length(G, source=s, target=t)
            
        except nx.NetworkXNoPath: 
            dis = float("inf")
        dis_m[s][t] = dis
        dis_m[t][s] = dis
    else:
        dis_m[s][t] = float("inf")
        dis_m[t][s] = float("inf")
```

```python
#clustering and filtering functions

def cluster_dataset(genes, dataset_name):
    #make matrix
    genes = list(set(genes))
    m = len(genes)
    samples = [[i,i] for i in range(m)]
    def get_dis(i, j):
        i, j = int(i[0]), int(j[0])
        a, b = genes[i], genes[j]
        dis = dis_m[a][b]
        if dis < float("inf"):return dis
        else:return 50

    Z = linkage(samples, method='single', optimal_ordering=True, metric=get_dis)
    plt.figure(figsize=(50, 10))
    plt.title('Hierarchical Clustering Dendrogram ' + dataset_name)
    plt.xlabel('sample index')
    plt.ylabel('distance')
    dendrogram(Z, leaf_rotation=90., leaf_font_size=6.
    )
    plt.show()
    return Z
def print_gene_in_cls(z, cutoff, cls_id, gene_set, data_set):
    ids = fcluster(z, cutoff, criterion='distance')
    genes = list(set(gene_set))
    string_ids = [genes[i] for i in range(len(ids)) if ids[i] == cls_id]
    res = pd.DataFrame(string_ids)
    res.columns = ['STRING_id']
    rs = pd.merge(res,STRING_mapping,on='STRING_id')
    #ids
    res = pd.merge(rs, datas[data_set], left_on = "gene", right_on = "#NearbyGene", how="left")['NearbyGeneSymbol']
    #print(res.to_string(index=False))
    return res
    
def create_cluster_count_data(z):
    cutoffs = range(1,20)
    data = {}
    for cutoff in cutoffs:
        labels = fcluster(z, cutoff, criterion='distance')
        unique, counts = np.unique(labels, return_counts=True)
        data[cutoff] = dict(zip(unique, counts))
    return data
```

```python
#clustering on antisc genes
genes = antisc_genes
antisc_genes_z = cluster_dataset(genes, "antisc_genes")
```

```python
#clustering on all sc genes
genes = antisc_genes + sc_genes
sc_z = cluster_dataset(genes, "sc_genes")
```

```python
#save sc filtering result
data = create_cluster_count_data(sc_z)
pd.DataFrame.from_dict(data, orient='index').replace({np.nan:0})[0:7].plot()
res = print_gene_in_cls(sc_z, 5, 1, sc_genes, "SC")
res = res.dropna()
res = sorted(list(set([ i for i in list(res) if i ])))
f = open('SC_gene_set.txt', 'w')
f.write('\n'.join(list(res)))
f.close()
```

```python
#clustering on all ms genes
genes = antims_genes +  ms_genes
ms_z = cluster_dataset(genes, "ms_genes")
```

```python
data = create_cluster_count_data(ms_z)
pd.DataFrame.from_dict(data, orient='index').replace({np.nan:0})[0:7].plot()
```

```python
#save MS filtering result
res = print_gene_in_cls(ms_z, 4, 1, ms_genes, "MS")
res = res.dropna()
res = sorted(list(set([ i for i in list(res) if i ])))
f = open('MS_gene_set.txt', 'w')
f.write('\n'.join(list(res)))
f.close()
```
